```python
import pandas as pd 


airports = pd.read_csv('airport_codes.csv')

airports.type.unique()

airports.head()

```

```python
values_to_drop = ['balloonport', 'heliport', 'seaplane_base', 'closed']

# Using boolean indexing to filter rows
df_filtered = airports[~airports['type'].isin(values_to_drop)]

df_filtered['Unnamed: 0'] = df_filtered['Unnamed: 0'] -1

df_filtered = df_filtered.rename(columns={'Unnamed: 0': ''})



# If you want to modify the original DataFrame in place, you can use the inplace parameter
# df.drop(df[df['type'].isin(values_to_drop)].index, inplace=True)
# Display the resulting DataFrame
df_filtered.type.unique()
df_filtered.head()
```

```python
df_filtered.to_csv('airportCodesTypeFiltered.csv', index=False)
```
